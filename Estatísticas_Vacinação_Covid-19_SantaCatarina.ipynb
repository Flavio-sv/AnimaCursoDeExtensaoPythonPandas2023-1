```python
# Dropando colunas que não serão utilizadas para análise.
df.drop(columns=['document_id', 'paciente_id', 'paciente_racaCor_codigo', 'paciente_endereco_coIbgeMunicipio', 'paciente_endereco_coPais', 'paciente_endereco_cep', 'paciente_nacionalidade_enumNacionalidade', 'estabelecimento_valor', 'estabelecimento_municipio_codigo', 'vacina_grupoAtendimento_codigo', 'vacina_categoria_codigo', 'vacina_fabricante_referencia', 'vacina_codigo'], inplace=True)
```

```python
# Agrupando e exibindo a contagem de pessoas vacinadas por município.
grouped = df.groupby(['municipio_nome', 'municipio_uf'])
result = grouped.count()
result_df = pd.DataFrame({'total_pessoas_vacinadas': result['paciente_idade']})
result_df.reset_index(inplace=True)

display(result_df)
```

```python
# Agrupando e exibindo a contagem de pessoas vacinadas por município, do mais vacinado para o menos vacinado.
grouped = df.groupby(['municipio_nome', 'municipio_uf'])
result = grouped.count()
result_df = pd.DataFrame({'total_pessoas_vacinadas': result['paciente_idade']})
result_df.reset_index(inplace=True)
result_df.sort_values(by='total_pessoas_vacinadas', ascending=False, inplace=True)
result_df.reset_index(drop=True, inplace=True)

display(result_df)
```

```python
# Agrupando os dados do dataframe original (df) com base na coluna 'vacina_descricao_dose'. 
# Em seguida, criando uma tabela pivot, que organiza os dados agrupados em uma tabela resumida, com as linhas representando 'municipio_nome' e 'municipio_uf', e as colunas representando cada dose de vacina diferente presente na coluna 'vacina_descricao_dose'. 
# O valor em cada célula da tabela corresponde ao número total de pessoas vacinadas com aquela dose específica.
grouped = df.groupby(['municipio_nome', 'municipio_uf', 'vacina_descricao_dose'])
result = grouped.count()
result_df = pd.DataFrame({'total_pessoas_vacinadas': result['paciente_idade']})
result_df = result_df.reset_index()
result_df = result_df.pivot_table(index=['municipio_nome', 'municipio_uf'], columns='vacina_descricao_dose', values='total_pessoas_vacinadas')
result_df.columns = result_df.columns.rename(None)
result_df = result_df.reset_index()
result_df.fillna(0, inplace=True)

display(result_df)
```

```python
# Somando todos os valores numéricos em cada coluna (que representa uma dose diferente de vacina) e exibindo o resultado final como um objeto do tipo Series (um tipo de estrutura de dados unidimensional do Pandas).
# Desta forma, é possível visualizar a quantidade total de doses de cada vacina agrupadas por descrição de dose.
result_df_sums = result_df.select_dtypes(include=['float']).sum()

display(result_df_sums)
```

```python
# Fazendo uma cópia da tabela "result_df" usando o método .copy() para garantir que não afetaremos a tabela original.
# Utilizando o .loc[] para selecionar as colunas que desejamos e agrupá-las para uma melhor organização.
# Calculando o total de doses aplicadas de cada tipo e somando as colunas selecionadas.
# Removendo as colunas antigas, exibimos o novo DataFrame resultante.
copy_segurança_df = result_df.copy()
copy_segurança_df['1ª_dose'] = copy_segurança_df.loc[:, ['1ª Dose', '1ª Dose Revacinação', 'Dose', 'Única']].sum(axis=1)
copy_segurança_df['2ª_dose'] = copy_segurança_df.loc[:, ['2ª Dose', '2ª Dose Revacinação']].sum(axis=1)
copy_segurança_df['3ª_dose'] = copy_segurança_df.loc[:, ['3ª Dose', '3ª Dose Revacinação']].sum(axis=1)
copy_segurança_df['4ª_dose'] = copy_segurança_df.loc[:, ['4ª Dose Revacinação']].sum(axis=1)
copy_segurança_df['5ª_dose'] = copy_segurança_df.loc[:, ['5ª Dose Revacinação']].sum(axis=1)
copy_segurança_df['dose_reforço'] = copy_segurança_df.loc[:, ['1º Reforço', '2º Reforço', '3º Reforço', 'Dose Adicional', 'Revacinação', 'Reforço']].sum(axis=1)
copy_segurança_df['dose_totais_aplicada'] = copy_segurança_df.loc[:, ['1ª Dose', '1ª Dose Revacinação', 'Dose', 'Única', '2ª Dose', '2ª Dose Revacinação', '3ª Dose', '3ª Dose Revacinação', '4ª Dose Revacinação', '5ª Dose Revacinação', '1º Reforço', '2º Reforço', '3º Reforço', 'Dose Adicional', 'Revacinação', 'Reforço']].sum(axis=1)
copy_segurança_df = copy_segurança_df.drop(columns=['1ª Dose', '2ª Dose', '3ª Dose','1ª Dose Revacinação', '1º Reforço', '2ª Dose Revacinação', '2º Reforço', '3ª Dose Revacinação', '3º Reforço', '4ª Dose Revacinação', '5ª Dose Revacinação', 'Dose', 'Dose Adicional', 'Reforço', 'Revacinação', 'Única'])

display(copy_segurança_df)
```

```python
# Nesse bloco, tem o mesmo objetivo que o anterior apenas com uma modificação que classifica os municípios que mais aplicaram a primeira dose de vacinação, utilizando o método .sort_values. 
# É possível fazer isso com todas as outras doses, basta mudar o valor da dose desejada para classificar.
copy_df = result_df.copy()
copy_df['1ª_dose'] = copy_df.loc[:, ['1ª Dose', '1ª Dose Revacinação', 'Dose', 'Única']].sum(axis=1)
copy_df['2ª_dose'] = copy_df.loc[:, ['2ª Dose', '2ª Dose Revacinação']].sum(axis=1)
copy_df['3ª_dose'] = copy_df.loc[:, ['3ª Dose', '3ª Dose Revacinação']].sum(axis=1)
copy_df['4ª_dose'] = copy_df.loc[:, ['4ª Dose Revacinação']].sum(axis=1)
copy_df['5ª_dose'] = copy_df.loc[:, ['5ª Dose Revacinação']].sum(axis=1)
copy_df['dose_reforço'] = copy_df.loc[:, ['1º Reforço', '2º Reforço', '3º Reforço', 'Dose Adicional', 'Revacinação', 'Reforço']].sum(axis=1)
copy_df['dose_totais_aplicada'] = copy_df.loc[:, ['1ª Dose', '1ª Dose Revacinação', 'Dose', 'Única', '2ª Dose', '2ª Dose Revacinação', '3ª Dose', '3ª Dose Revacinação', '4ª Dose Revacinação', '5ª Dose Revacinação', '1º Reforço', '2º Reforço', '3º Reforço', 'Dose Adicional', 'Revacinação', 'Reforço']].sum(axis=1)
copy_df = copy_df.drop(columns=['1ª Dose', '2ª Dose', '3ª Dose','1ª Dose Revacinação', '1º Reforço', '2ª Dose Revacinação', '2º Reforço', '3ª Dose Revacinação', '3º Reforço', '4ª Dose Revacinação', '5ª Dose Revacinação', 'Dose', 'Dose Adicional', 'Reforço', 'Revacinação', 'Única'])
copy_df.sort_values(by='1ª_dose', ascending=False, inplace=True)
copy_df.reset_index(drop=True, inplace=True)

display(copy_df)
```

```python
# exibindo a média de idade dos pacientes vacinados por município. 
# Selecionamos as colunas 'paciente_idade', 'municipio_nome' e 'municipio_uf' do dataframe original 'df' e armazena em 'vacinados_idaMedia_df'. 
# Agrupamos os dados por 'municipio_nome' e 'municipio_uf' e calcula a média da idade dos pacientes vacinados, armazenando o resultado em 'vacinados_grouped_df'. 
# Por fim, alteramos o nome da coluna 'paciente_idade' para 'idade_media_vacinado_porMunicipio' e exibimos o resultado final.
vacinados_idaMedia_df = df[['paciente_idade', 'municipio_nome', 'municipio_uf']]
vacinados_grouped_df = vacinados_idaMedia_df.groupby(['municipio_nome', 'municipio_uf']).agg({'paciente_idade': 'mean'})
vacinados_grouped_df = vacinados_grouped_df.reset_index()
vacinados_grouped_df.rename(columns={'paciente_idade': 'idade_media_vacinado_porMunicipio'}, inplace=True)

display(vacinados_grouped_df)
```
